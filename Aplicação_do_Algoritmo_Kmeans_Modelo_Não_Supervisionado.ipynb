```python
# clustering dataset
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])

plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset')
plt.scatter(x1, x2)
plt.show()

# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']

# KMeans algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])

plt.show()
```

```python
#Dataset IrisPython

import pandas as pd # Import o pandas para manipular os dados.

iris = pd.read_csv("iris(1).csv") # Carregar os Dados e colocar no Data Frame iris.

iris.head() # Mostrar as primeiras linhas 5 do data frame
```

```python
# Remover a coluna Espécie no final do data frame.
X = iris.iloc[:, 0:4].values
X
```

```python
# Instanciar o Objeto do KMeans
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
# Treinando o modelo de ML (Kmeans)
kmeans.fit(X)
```

```python
# Exibir a quantidade de Centroides identificados.
kmeans.cluster_centers_
```

```python
# Exibir a distância entre os pontos centrais.
distance = kmeans.fit_transform(X)
distance
```

```python
# Exibindo os resultados do Cluster Completo.
labels = kmeans.labels_
labels
```

```python
# Método Elbow -> O método Elbow se trata de uma técnica interessante para encontrar o valor ideal do parâmetro k.
# Basicamente o que o método faz é testar a variância dos dados em relação ao número de clusters.
```

```python
from sklearn.cluster import KMeans
wcss = []
i = 0 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```
